```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("Exam_Score_Dataset.csv")
df
```

## Age

```python
bins = range(df['age'].min(), df['age'].max() + 1, 1)
plt.hist(df['age'].tolist(), bins = bins, edgecolor='black', linewidth=1.2)
plt.title('Liczba uczniów w poszczególnej grupie wiekowej')
plt.xlabel('Grupa wiekowa')
plt.ylabel('Liczba uczniów')
```

## Gender

```python
labels = df['gender'].unique().tolist()
y = [df[df['gender'] == gender].shape[0] for gender in labels]
plt.pie(y, labels=labels, autopct='%1.1f%%')
plt.title("Rozkład płciowy uczniów")
```

## Course

```python
labels = df['course'].unique().tolist()
y = [df[df['course'] == course].shape[0] for course in labels]
plt.pie(y, labels=labels, autopct='%1.1f%%')
plt.title("Odsetek uczniów realizujaych konkrente programy nauczania")
```

## Internet access

```python
labels = df['internet_access'].unique().tolist()
y = [df[df['internet_access'] == row].shape[0] for row in labels]
plt.pie(y, labels=labels, autopct='%1.1f%%')
plt.title("Dostęp do internetu")
```

```python
print(f'Mediana uczniów mających dostęp do Internetu {np.median(df[df['internet_access'] == 'yes']['exam_score'])}')
print(f'Mediana uczniów mających dostęp do Internetu {np.median(df[df['internet_access'] == 'no']['exam_score'])}')
```

## Sleep Hours

```python
labels = ['less than 6h', 'more than 6h']
y = [df[df['sleep_hours'] < 6].shape[0], df[df['sleep_hours'] >= 6].shape[0] ]
plt.pie(y, labels=labels, autopct='%1.1f%%')
plt.title("Czas snu uczniów")
```

```python
print(f'Mediana uczniów śpiąych mniej niż 6 godzin {np.median(df[df['sleep_hours'] < 6]['exam_score'])}')
print(f'Mediana uczniów śpiących 6 lub więcej godzin {np.median(df[df['sleep_hours'] >= 6]['exam_score'])}')
```

## Sleep Quality

```python
labels = df['sleep_quality'].unique().tolist()
y = [df[df['sleep_quality'] == row].shape[0] for row in labels]
plt.pie(y, labels=labels, autopct='%1.1f%%')
plt.title("Jakość snu")
```

## Study method

```python
labels = df['study_method'].unique().tolist()
y = [df[df['study_method'] == row].shape[0] for row in labels]
plt.pie(y, labels=labels, autopct='%1.1f%%')
plt.title("Sposób nauki")
```

## Facility Rating

```python
labels = df['facility_rating'].unique().tolist()
y = [df[df['facility_rating'] == row].shape[0] for row in labels]
plt.pie(y, labels=labels, autopct='%1.1f%%')
plt.title("Warunki do nauki")
```

```python
labels = df['exam_difficulty'].unique().tolist()
y = [df[df['exam_difficulty'] == row].shape[0] for row in labels]
plt.pie(y, labels=labels, autopct='%1.1f%%')
plt.title("Poziom trudności egzaminu")
```
